```python
import sqlite3
import numpy as np
import pandas as pd
conn = sqlite3.connect('test.db')
```

```python
patient = pd.read_csv('./csv/patients.csv')
```

```python
patientid = []
for p in patient['Id']:
    patientid.append(p)
```

```python
pname= []
pincome = []
pexpense = []
pcoverage = []
for p in patient['FIRST']:
    pname.append(p)
for p in patient['INCOME']:
    pincome.append(p)
for p in patient['HEALTHCARE_EXPENSES']:
    pexpense.append(p)
for p in patient['HEALTHCARE_COVERAGE']:
    pcoverage.append(p)
```

```python
print((patientid[1]), (pname[1]),(pincome[1]),(pexpense[1]),(pcoverage[1]))
```

```python
length = len(patientid)
for i in range(1,length):
    conn.execute(f'''insert into patient values('{patientid[i]}','{pname[i]}',{pincome[i]},{pexpense[i]},{pcoverage[i]});''')


```
